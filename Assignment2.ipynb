```python
import numpy as np
```

```python
ytrue  =np.array([41,45,49,47,44])
ypred =np.array([43.6,44.4,45.2,46,46.8])
```

```python
# Mean Squared Error without using numpy
def mean_squared_error(ytrue,ypred):
  err = 0
  for yt, yp in zip(ytrue,ypred):
    err+=(yt-yp)**2
  print("Total Error",err)
  mse = (err/len(ytrue))
  return mse

me = mean_squared_error (ytrue,ypred)
print("Mean squared error is :",me)
```

```python
# implement mean square error using numpy
np.mean((ytrue-ypred)**2)
```
